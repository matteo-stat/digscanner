## load with opencv

# border detection

```python
tmp = img.copy()
tmp = tmp[:int(height*0.08)]
gray = cv.cvtColor(tmp, cv.COLOR_BGR2GRAY)
blur = cv.GaussianBlur(gray, ksize=(21, 21), sigmaX=0)
ret, th = cv.threshold(blur, thresh=0, maxval=255, type=cv.THRESH_BINARY + cv.THRESH_OTSU)

plt.imshow(th, cmap='gray', vmin=0, vmax=255)
deltas = np.diff(th.sum(axis=1))
y1 = np.argmax(deltas > 0)
y2 = len(deltas) - np.argmax(deltas[::-1] > 0) - 1
# plt.plot([0, tmp.shape[1]], [y1, y2], color='red', linestyle='-', linewidth=1)
# plt.plot(x, y, marker="o", markersize=1, color="red")
plt.show()
```

```python
deltas = np.diff((th < 255).sum(axis=1))
# angle_dir = -1 if np.sum(np.diff(deltas) > 0) > np.sum(np.diff(deltas) < 0) else 1
y = np.where(deltas)[0]
x = np.arange(0, len(y), 1)
A = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[0]
m, c, math.atan(m)
```

```python
plt.plot(x, y, 'o', label='Original data', markersize=10)
plt.plot(x, m*x + c, 'r', label='Fitted line')
plt.legend()
plt.show()
```

```python
res = np.abs(y - (m*x + c))
outlier_th = np.percentile(res, q=95, method='closest_observation')
outlier_th
```

```python
y = y[res <= outlier_th]
x = np.arange(0, len(y), 1)
A = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[0]
m, c, math.atan(m)
```

```python
plt.plot(x, y, 'o', label='Original data', markersize=10)
plt.plot(x, m*x + c, 'r', label='Fitted line')
plt.legend()
plt.show()
```

```python
center = (width/2, height/2)
angle = 0.5*math.atan(m)
angle = 0
scale = 1.0
M = cv.getRotationMatrix2D(center, angle, scale)
rotated = cv.warpAffine(img, M, (width, height))

plt.imshow(cv.cvtColor(rotated, cv.COLOR_BGR2RGB), interpolation='nearest')
plt.plot([0, rotated.shape[1]], [c, c], color='red', linestyle='-', linewidth=1)
plt.show()
```
